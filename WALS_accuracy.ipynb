```python
import pandas as pd
import numpy as np
```

```python
ratings_headers = ['user_id', 'movie_id', 'ratings', 'timestamp']
ratings = pd.read_table('u.data', sep='\t', header=None, names=ratings_headers)
ratings.head()
```

```python
rp = ratings.pivot_table(columns=['movie_id'],index=['user_id'],values='ratings')
rp.head()
```

```python
rp = rp.fillna(0); # Replace NaN
rp.head()
```

```python
Q = rp.values
```

```python
W = Q>0.1
W[W == True] = 1
W[W == False] = 0
# To be consistent with our Q matrix
W = W.astype(np.float64, copy=False)
```

```python
lambda_ = 0.06
n_factors = 15
m, n = Q.shape
n_iterations = 1000
```

```python
X = np.random.rand(m, n_factors) 
Y = np.random.rand(n_factors, n)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
errors = []
diff_error = 10
for ii in range(n_iterations):
    for u, Wu in enumerate(W):
        X[u] = np.linalg.solve(np.dot(Y, np.dot(np.diag(Wu), Y.T)) + lambda_ * np.eye(n_factors),np.dot(Y, np.dot(np.diag(Wu), Q[u].T))).T
    for i, Wi in enumerate(W.T):
        Y[:,i] = np.linalg.solve(np.dot(X.T, np.dot(np.diag(Wi), X)) + lambda_ * np.eye(n_factors),np.dot(X.T, np.dot(np.diag(Wi), Q[:, i])))
    errors.append(mean_squared_error(Q,np.dot(X,Y)))
    print('{}th iteration is completed'.format(ii))
    if(ii>1):
        diff_error = abs(errors[ii-2] - errors[ii-1])
    if diff_error < 0.0000001:
        break;
```

```python
now_ratings = np.dot(X,Y)
```

```python
mse = mean_squared_error(Q,now_ratings)
```

```python
print("MSE =  %f" %mse)
```

```python
import math
print(math.sqrt(mse))
```
