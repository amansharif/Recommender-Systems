```python
import numpy as np
import pandas as pd
```

```python
x = np.array([[5,0,4,0,3,3,0],
            [0,4,0,0,4,0,2],
            [3,0,3,3,0,0,3],
            [0,4,0,5,0,0,0],
             [0,0,4,0,0,3,0]])
```

```python
Data = pd.DataFrame(data=x, index=['user_1','user_2','user_3','user_4','user_5'], 
                    columns=['movie_1','movie_2','movie_3','movie_4','movie_5','movie_6','movie_7'])
Data.replace(0,np.NaN)
```

```python
for_user_based = Data
for_movie_based = Data.T
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x = cosine_similarity(for_user_based)
user_similarity = pd.DataFrame(data=x, index=['user_1','user_2','user_3','user_4','user_5'], 
                           columns=['user_1','user_2','user_3','user_4','user_5'])
user_similarity.head()
```

```python
y = cosine_similarity(for_movie_based)
movie_similarity = pd.DataFrame(data=y, index=['movie_1','movie_2','movie_3','movie_4','movie_5','movie_6','movie_7'],
                                columns=['movie_1','movie_2','movie_3','movie_4','movie_5','movie_6','movie_7'])
movie_similarity.head(7)
```
